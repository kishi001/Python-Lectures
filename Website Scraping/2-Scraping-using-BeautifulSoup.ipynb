```python
import requests #request is used to access web pages
page = requests.get("https://en.wikipedia.org/wiki/Main_Page")
contents = page.content
```

```python
contents
```

```python
from bs4 import BeautifulSoup as BS #BS4 is a parsing library
soup = BS(contents, 'html.parser')
soup.find_all('a') #let's scrape all 'a' tags
```

```python
for i in soup.find_all('a'): #this is to get the texts that are hyperlinked.
    print (i.text) 
```
